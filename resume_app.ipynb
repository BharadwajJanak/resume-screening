```python
import streamlit as st
import pickle
import re
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
STOPWORDS = set(stopwords.words("english"))
```

```python
tfidf = pickle.load(open("tfidf.pkl", "rb"))
model = pickle.load(open("model.pkl", "rb"))
```

```python
def clean_text(text):
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z ]", "", text)
    text = text.lower()
    return " ".join(w for w in text.split() if w not in STOPWORDS)
```

```python
def predict_resume(text):
    cleaned = clean_text(text)
    vector = tfidf.transform([cleaned])
    prediction = model.predict(vector)[0]
    confidence = model.predict_proba(vector).max()
    return prediction, confidence
```

```python
st.set_page_config(page_title="Resume Screening ML App", layout="wide")

st.title("ðŸ“„ Resume Screening using Machine Learning")
st.markdown("Paste resume text below to predict the job category.")
```

```python
resume_text = st.text_area("Resume Text", height=250)

if st.button("Predict Category"):
    if resume_text.strip() == "":
        st.warning("Please enter resume text.")
    else:
        category, confidence = predict_resume(resume_text)
        st.success(f"**Predicted Category:** {category}")
        st.info(f"**Confidence:** {confidence:.2%}")
```
